```python
# Experiment 8 - Vertical, Horizontal and Diagonal Edges
from skimage import io
import numpy as np
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from scipy import signal
```

```python
image = io.imread("baboon.png")
image = rgb2gray(image)
plt.figure()
io.imshow(image, cmap='gray')
plt.title("Original Image")
```

```python
# Horizontal edges
sh = image.shape
row = sh[0]
col = sh[1]

horz = [[1,2,1],[0,0,0],[-1,-2,-1]]
horz_image = signal.convolve2d(image,horz, mode='same')

for r in range(0,row):
    for c in range(0,col):
        if horz_image[r][c] < 0:
            horz_image[r][c] = 0
        else:
            continue

plt.figure()
io.imshow(horz_image, cmap='gray')
plt.title('Horizontal Image')
```

```python
# Vertical edges
sh = image.shape
row = sh[0]
col = sh[1]

vert = [[1,0,-1],[2,0,-2],[1,0,-1]]
vert_image = signal.convolve2d(image,vert, mode='same')

for r in range(0,row):
    for c in range(0,col):
        if vert_image[r][c] < 0:
            vert_image[r][c] = 0
        else:
            continue

plt.figure()
io.imshow(vert_image, cmap='gray')
plt.title('Vertical Image')
```

```python
# Forward Diagonal edges
sh = image.shape
row = sh[0]
col = sh[1]

fd = [[0,1,2],[-1,0,1],[-2,-1,0]]
fd_image = signal.convolve2d(image,fd, mode='same')

for r in range(0,row):
    for c in range(0,col):
        if fd_image[r][c] < 0:
            fd_image[r][c] = 0
        else:
            continue
```

```python
# Backward Diagonal edges
sh = image.shape
row = sh[0]
col = sh[1]

bd = [[0,-1,-2],[1,0,-1],[2,1,0]]
bd_image = signal.convolve2d(image,bd, mode='same')

for r in range(0,row):
    for c in range(0,col):
        if bd_image[r][c] < 0:
            bd_image[r][c] = 0
        else:
            continue


plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
io.imshow(fd_image, cmap='gray')
plt.title('Forward Diagonal Image')

plt.subplot(1,2,2)
io.imshow(bd_image, cmap='gray')
plt.title('Backward Diagonal Image')
```

```python
d_image = fd_image + bd_image

plt.figure()
io.imshow(d_image, cmap='gray')
plt.title('Diagonal Image')
```

```python
# Enhance all types of edges

edge = [[1,1,1],[1,-8,1],[1,1,1]]
edge_image = signal.convolve2d(image,edge, mode='same')

for r in range(0,row):
    for c in range(0,col):
        if edge_image[r][c] < 0:
            edge_image[r][c] = 0
        else:
            continue

final_image = image + edge_image

plt.figure(figsize=(10,10))

plt.subplot(1,3,1)
io.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1,3,2)
io.imshow(edge_image, cmap='gray')
plt.title('Laplacian Image')

plt.subplot(1,3,3)
io.imshow(final_image, cmap='gray')
plt.title('Final Image')
```

Conclusion: -

To enhance horizontal, vertical and diagonal edges of the image sobel filter is used.

Enhanced image shows highlighted vertical, horizontal and diagonal edges.

Laplacian operator enhances all types of edges for the given image.

For overall enhancement of the image, image is added with the enhanced edges.
