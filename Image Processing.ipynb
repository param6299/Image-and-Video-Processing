```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Change the format
im = Image.open('peppers.png')
plt.subplot(1,2,1)
plt.imshow(im)
plt.title('PNG Image')
print(im.mode)
im.save('peppers.jpg')

im_j = Image.open('peppers.jpg')
plt.subplot(1,2,2)
plt.imshow(im_j)
plt.title('JPG Image')
```

```python
from skimage.io import imread, imsave, imshow
from skimage import color
```

```python
# Change the color space
im = imread('pool.png')
im_hsv = color.rgb2hsv(im)

plt.figure(figsize=(7,7))
plt.subplot(2,2,1)
plt.imshow(im)
plt.title('Original Image')

im_h = im_hsv[:,:,0]
im_s = im_hsv[:,:,1]
im_v = im_hsv[:,:,2]

plt.subplot(2,2,2)
plt.imshow(im_h, cmap='gray')
plt.title('Hue part of Original Image')

plt.subplot(2,2,3)
plt.imshow(im_s, cmap='gray')
plt.title('Saturation part of Original Image')

plt.subplot(2,2,4)
plt.imshow(im_v, cmap='gray')
plt.title('Brightness part of Original Image')
```

```python
# Image Blending
# Start with one image and end with other image
from skimage.transform import resize
import numpy as np
```

```python
im1 = imread('dhoni.jpg')
im1 = im1/255

im2 = imread('ViratKohli.jpg')
im2 = im2/255

sh = im2.shape
im1=resize(im1,(sh[0],sh[1],3))

plt.figure(figsize=(20,20))

i=1

for alpha in np.linspace(0,1,20):
    plt.subplot(4,5,i)
    im = ((1-alpha)*im1)+(alpha*im2)
    plt.imshow(im)
    i=i+1
```
