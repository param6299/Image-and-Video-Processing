```python
# Experiment 3 - Point Operation
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
image = io.imread('lenna_grey.jpg')
io.imshow(image)
```

```python
# Scaling the brightness of the image
def scale(x,a):
    y=x*a
    return y

image_scale = image.copy()
sh=image_scale.shape
# scaling factor is 0.5
for row in range(0,sh[0]):
    for col in range(0,sh[1]):
        temp1 = image[row][col]
        temp2 = scale(temp1,1.5)
        image_scale[row][col] = temp2
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow(image)
plt.title('Original Image')

plt.subplot(1,2,2)
io.imshow(image_scale)
plt.title('Scaled Image')
```

```python
# Scaling the brightness of the image
def scale(x,a):
    y=x+a
    return y

image_scale = image.copy()
sh=image_scale.shape
# scaling factor is 0.5
for row in range(0,sh[0]):
    for col in range(0,sh[1]):
        temp1 = image[row][col]
        temp2 = scale(temp1,100)
        image_scale[row][col] = temp2
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow(image)
plt.title('Original Image')

plt.subplot(1,2,2)
io.imshow(image_scale)
plt.title('Brightened Image')
```

```python
# Thresholding without background

image_gray=rgb2gray(image)
sh=image_gray.shape
th=input("Enter threshold value:")
th=int(th)

image_th = image_gray.copy()

for row in range(0,sh[0]):
    for col in range(0,sh[1]):
        if image_th[row][col]*255 >= th:
            image_th[row][col] = 255
        else:
            image_th[row][col] = 0
            
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow(image_gray)
plt.title('Gray Image')

plt.subplot(1,2,2)
io.imshow(image_th,cmap='gray')
plt.title('Threshold Image')
```

```python
# Thresholding with background

image_gray=rgb2gray(image)
sh=image_gray.shape
th=input("Enter threshold value:")
th=int(th)
image_gray = image_gray*255
image_th = image_gray.copy()

for row in range(0,sh[0]):
    for col in range(0,sh[1]):
        if image_gray[row][col]>= th:
            image_th[row][col] = 255
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow(image)
plt.title('Gray Image')

plt.subplot(1,2,2)
io.imshow(image_th,cmap='gray')
plt.title('Threshold Image')
```
