```python
# Hadamard Transform  for Image
from scipy.linalg import hadamard
import numpy as np
import random
```

```python
# Generate Hadamard matrix of order N
N = input("Enter order of Hadamard matrix-")
N = int(N)
H = hadamard(N)
print(H)
```

```python
# Generate original matrix
mat = np.random.randint(9,size=(N,N))

# Transformed matrix is F = H*f*H'
temp = np.dot(H,mat)
trans_mat = np.dot(temp,np.transpose(H))
trans_mat
```

```python
# inverse Hadamard transform
# f = H'*F*H
temp_f = np.dot(np.transpose(H),trans_mat)
inv_mat = np.dot(temp_f,H)
inv_mat = inv_mat//(N*N)
print('Original Matrix- \n',mat)
print('Transformed Matrix- \n',trans_mat)
print('Inverse of Hadamard Matrix- \n',inv_mat)
```

```python
# Hadamard Transform of Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.linalg import hadamard
```

```python
# Definition for Hadamard Transformation of 8x8 subimage
def had2(f):
    H = hadamard(8)
    F = np.dot(np.dot(H,f),np.transpose(H))
    return F

# Definition for Inverse Hadamard Transformation of 8x8 subimage
def had2inv(F):
    H = hadamard(8)
    f = (np.dot(np.dot(np.transpose(H),F),H))//8
    return f
```

```python
image = cv2.imread('roof.tif',0)
sh=image.shape
row=sh[0]
col=sh[1]

plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Hadamard Transform of original image

image_hada = np.zeros((row,col),dtype=int)

for r in range(row//8):
    for c in range(col//8):
        image_hada[r*8:(r+1)*8,c*8:(c+1)*8] = had2(image[r*8:(r+1)*8,c*8:(c+1)*8])

plt.subplot(1,3,2)
plt.imshow(image_hada, cmap='gray')
plt.title('Hadamard Transformed Image')

# Inverse of Hadamard Transform of original image

image_inv = np.zeros((row,col),dtype=int)

for r in range(row//8):
    for c in range(col//8):
        image_inv[r*8:(r+1)*8,c*8:(c+1)*8] = had2inv(image_hada[r*8:(r+1)*8,c*8:(c+1)*8])

plt.subplot(1,3,3)
plt.imshow(image_inv, cmap='gray')
plt.title('Inverse Image')
```

Conclusion: - 

Elements of Hadamard matrix are either +1 or -1.

Hadamard matrix is symmetric and has real values

Hadamard matrix can be applied on matrices of multiples of 2.

It is observed that original image and inverse of transformed image is same. Therefore hadamard matrix can be used for signal processing
