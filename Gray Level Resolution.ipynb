```python
# Experiment 1 - Gray level Resolution
from skimage import data
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
image = data.astronaut()
io.imshow(image)
image_gray = rgb2gray(image)
#io.imshow(image_gray)
#print(image_gray)
image_gray = image_gray*255
```

```python
image_reso = image_gray.copy()
b=3
num_levels = 2**b
step_size = int(256/num_levels)
sh = image_reso.shape
```

```python
for r in range(0,sh[0]):
    for c in range(0,sh[1]):
        temp = image_reso[r][c]
        for  i in range(0,num_levels):
            if temp >= i*step_size and temp<(i+1)*step_size:
                image_reso[r][c] = i*step_size
                break
```

```python
plt.figure()
plt.figure(figsize=(7,7))

plt.subplot(1,2,1)
io.imshow(image_reso, cmap='gray')
plt.title('Image with low resolution')

plt.subplot(1,2,2)
io.imshow(image_gray, cmap='gray')
plt.title('Original image')
```
