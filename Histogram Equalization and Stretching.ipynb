```python
# Experiment 5

from scipy import misc
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = misc.face()
plt.figure()
io.imshow(img)
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
ax = plt.hist(img[:,:,0].ravel(),bins=256,color='r')# ravel is used to convert 2D array into 1D array
plt.xlabel('Intensity')
plt.ylabel('No of pixels')
plt.title('Histogram for Red Plane')

plt.subplot(1,3,2)
ax = plt.hist(img[:,:,1].ravel(),bins=256,color='g')# ravel is used to convert 2D array into 1D array
plt.xlabel('Intensity')
plt.ylabel('No of pixels')
plt.title('Histogram for Green Plane')

plt.subplot(1,3,3)
ax = plt.hist(img[:,:,2].ravel(),bins=256,color='b')# ravel is used to convert 2D array into 1D array
plt.xlabel('Intensity')
plt.ylabel('No of pixels')
plt.title('Histogram for Blue Plane')
```

```python
img1 = io.imread('pollen_very_dark.tif')
#io.imshow(img1)
sh = img1.shape
row = sh[0]
col = sh[1]

# no of pixels
numb_pix = row*col

# histogram
imghist,bins = np.histogram(img1.ravel(),256)

# cumulative sum for histogram
cdf = imghist.cumsum()
cdf_modified = cdf*255/numb_pix
s = cdf_modified.astype(int)

im_eq = img1.copy()
for r in range(0,row):
    for c in range(0,col):
        temp=img1[r][c]
        im_eq[r][c] = s[temp]
        
plt.figure(figsize=(7,7))

plt.subplot(2,2,1)
io.imshow(img1)
plt.title('Original image')

plt.subplot(2,2,2)
io.imshow(im_eq)
plt.title('Equalized image')

plt.subplot(2,2,3)
ax = plt.hist(img1.ravel(),bins=256)
plt.title('Histogram of Very Dark image')

plt.subplot(2,2,4)
ax = plt.hist(im_eq.ravel(),bins=256)
plt.title('Histogram of Equalized image')
```

Conclusion-

After equalizing these images, histogram is redistributed in the range 0 to 255. Therefore, all these equalized images have pixel intensity from 0 to 255.

```python
# Histogram streching
# streches minimum and maximum pixel intensity in original image to the range, 0 to 255
# pixel intensity r, in original image is converted to s in streched

image = io.imread('pollen_very_dark.tif')
sh = image.shape
row=sh[0]
col=sh[1]
s_min=0
s_max=255

r_min = np.min(np.min(image))
r_max = np.max(np.max(image))

image_str = image.copy()
for r in range(0,row):
    for c in range(0,col):
        temp1 = image[r][c]
        temp2 = ((s_max-s_min)/(r_max-r_min))*(temp1-r_min)+s_min
        image_str[r][c] = temp2
        
plt.figure(figsize=(7,7))

plt.subplot(1,2,1)
io.imshow(image)
plt.title('Original image')

plt.subplot(1,2,2)
io.imshow(image_str)
plt.title('Streched image')
```

Conclusion-

Histogram streching reduces minimum pixel intensity of original image to 0 and increases maximum pixel intensity to 255.

Intermediate pixel intensity of original image are increased proportionately.

In histogram equalization no of pixels with various intensity values are redistributed.

For histogram streching the pixel intensities are transformed to new values to cover the entire range, they are not redistributed.
